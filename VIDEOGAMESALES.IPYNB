WHICH NUMBER OF GAMES HAVE BEEN RELEASED MOST IN EACH YEAR?

IT SHOWS ACTION GAMES TOOK THE TOP 3 GAME GENRE RELEASED IN THE YEARS 2009, 2012 AND 2015 RESPECTIVELY

```python
year_max_df = df.groupby(['Year', 'Genre']).size().reset_index(name = 'Count')

year_max_idx = year_max_df.groupby(['Year'])['Count'].transform(max) == year_max_df['Count']

year_max_genre = year_max_df[year_max_idx].reset_index(drop = True)

year_max_genre = year_max_genre.drop_duplicates(subset = ['Year', 'Count'], keep= 'last').reset_index(drop=True)

genre = year_max_genre['Genre'].values
genre
```

```python
plt.figure(figsize = (20, 8))
graph = sns.barplot(x='Year', y = 'Count', data = year_max_genre)
index = 0
for value in year_max_genre['Count'].values:
    graph.text(index, value + 5, str(genre[index] + '-------' + str(value)), color = '#000', size = 14, rotation = 90, ha = 'center')
    index += 1
plt.xticks(rotation = 90)
plt.show()
```

WHICH GENRE GAME HAS THE MOST SALES IN A SINGLE YEAR

MORE SALES WERE MADE IN 2009...139.36MILLION AND 2008 WITH 136.39MILLION SALES MADE FOLLOWED BY SALES MADE FROM SPORTS IN 2006....136.16MILLION

```python
year_sale_dx = df.groupby(by=['Year', 'Genre'])['Global_Sales'].sum().reset_index()

year_sale = year_sale_dx.groupby(by = ['Year'])['Global_Sales'].transform(max) == year_sale_dx['Global_Sales']

year_sale_max = year_sale_dx['Global_Sales']

year_sale_max = year_sale_dx[year_sale].reset_index(drop = True)

genre = year_sale_max['Genre']
```

```python
plt.figure(figsize = (20, 8))
graph = sns.barplot(x='Year', y = 'Global_Sales', data = year_sale_max)
index = 0
for value in year_sale_max['Global_Sales'].values:
    graph.text(index, value + 5, str(genre[index] + '-------' + str(round(value, 2))), color = '#000', size = 14, rotation = 90, ha = 'center')
    index += 1
plt.xticks(rotation = 90)
plt.show()
```

SALES COMPARISON BY GENRE

```python
comp_genre = df[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

comp_map = comp_genre.groupby(by = ['Genre']).sum()
comp_map

plt.figure(figsize = (20, 10))
sns.set(font_scale=1)
sns.heatmap(comp_map, annot = True, fmt = '.1f')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

HERE IT SHOWS THAT NA_SALES HAS THE HIGHEST SALES AND ACTRION, SHOOTER, SPORTS ARE TOPPING THEIR CHARTS.....SEEMS THY LOVE VIDEO GAMES MORE THAN EU AND OTHER COUNTRIES

```python
comp_table = comp_map.reset_index()

comp_table = pd.melt(comp_table, id_vars = ['Genre'], value_vars = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name = 'Sales_Area', value_name = 'Sale_Price')
```

```python
plt.figure(figsize = (22, 10))
graph = sns.barplot(x = 'Genre', y = 'Sale_Price', hue = 'Sales_Area', data = comp_table)

for x in graph.containers:
    graph.bar_label(x)
plt.xticks(rotation = 90)
plt.show()
```

TOTAL REVENUE BY REGION

NA_SALES LEAD THE TOTAL SALE MADE BY 4327.65MILLION SALES FOLLWED BT EU_SALES WHICCH IS 2406.69MILLION SALES MADE
